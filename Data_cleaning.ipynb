Tables

```python
def ready(table):
    table = table.T
    table.columns = table.iloc[0]
    table.reset_index(inplace=True)
    table.rename(columns={1:'Date'},inplace=True)
    table.drop([0],axis=0,inplace=True)
    return table
```

```python
tbl_pl = ready(tbl_pl)
```

```python
tbl_pl.rename(columns={0:'Date'},inplace=True)
```

```python
tbl_pl = tbl_pl[:10]
```

```python
tbl_Balance_Sheet = ready(tbl_Balance_Sheet)
```

```python
lst=[]
count =1
for i in tbl_Balance_Sheet.columns:
    if i=='Total' and count==1:
        lst.append('Total_Liabilities')
        count += 1
    elif i=='Total' and count ==2:
        lst.append('Total_Assets')
    else:
        lst.append(i)

tbl_Balance_Sheet.columns = lst
```

```python
tbl_Cash_Flow = ready(tbl_Cash_Flow)
```

```python
dtl_bs = ready(dtl_bs)
```
